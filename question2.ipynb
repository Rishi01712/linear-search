```python
import random
import matplotlib.pyplot as plt
import numpy as np

def linear_search_with_comparisons(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            return i, comparisons
    return -1, comparisons

def generate_random_array(n):
    return list(range(1, n + 1))

def simulate_average_case(arr, trials=1000):
    total_comparisons = 0
    n = len(arr)

    for _ in range(trials):
        target = random.choice(arr)
        _, comparisons = linear_search_with_comparisons(arr, target)
        total_comparisons += comparisons

    return total_comparisons / trials

def analyze_linear_search_comparisons():
    sizes = [100, 500, 1000, 5000, 10000]

    best_case_comparisons = []
    average_case_comparisons = []
    worst_case_comparisons = []

    print("LINEAR SEARCH COMPARISON ANALYSIS")
    print("=" * 60)
    print(f"{'Size':<8} {'Best Case':<12} {'Average Case':<15} {'Worst Case':<12}")
    print("-" * 60)

    for n in sizes:
        arr = generate_random_array(n)
        random.shuffle(arr)

        best_target = arr[0]
        _, best_comparisons = linear_search_with_comparisons(arr, best_target)
        best_case_comparisons.append(best_comparisons)

        avg_comparisons = simulate_average_case(arr, trials=1000)
        average_case_comparisons.append(avg_comparisons)

        worst_target = max(arr) + 1000
        _, worst_comparisons = linear_search_with_comparisons(arr, worst_target)
        worst_case_comparisons.append(worst_comparisons)

        print(f"{n:<8} {best_comparisons:<12} {avg_comparisons:<15.2f} {worst_comparisons:<12}")

    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.plot(sizes, best_case_comparisons, 'bo-', label='Best Case', linewidth=2, markersize=8)
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Comparisons')
    plt.title('Best Case: O(1)')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(sizes, average_case_comparisons, 'go-', label='Average Case', linewidth=2, markersize=8)
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Comparisons')
    plt.title('Average Case: O(n/2)')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.subplot(2, 2, 3)
    plt.plot(sizes, worst_case_comparisons, 'ro-', label='Worst Case', linewidth=2, markersize=8)
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Comparisons')
    plt.title('Worst Case: O(n)')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.subplot(2, 2, 4)
    plt.plot(sizes, best_case_comparisons, 'bo-', label='Best Case O(1)', linewidth=2, markersize=6)
    plt.plot(sizes, average_case_comparisons, 'go-', label='Average Case O(n/2)', linewidth=2, markersize=6)
    plt.plot(sizes, worst_case_comparisons, 'ro-', label='Worst Case O(n)', linewidth=2, markersize=6)
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Comparisons')
    plt.title('Combined Comparison Analysis')
    plt.grid(True, alpha=0.3)
    plt.legend()

    plt.tight_layout()
    plt.show()

    print("\n" + "=" * 60)
    print("THEORETICAL vs EMPIRICAL ANALYSIS")
    print("=" * 60)

    print(f"{'Size':<8} {'Theoretical':<25} {'Empirical':<25}")
    print(f"{'':8} {'Best':<6} {'Avg':<9} {'Worst':<6} {'Best':<6} {'Avg':<9} {'Worst':<6}")
    print("-" * 60)

    for i, n in enumerate(sizes):
        theoretical_best = 1
        theoretical_avg = n // 2
        theoretical_worst = n

        empirical_best = best_case_comparisons[i]
        empirical_avg = average_case_comparisons[i]
        empirical_worst = worst_case_comparisons[i]

        print(f"{n:<8} {theoretical_best:<6} {theoretical_avg:<9} {theoretical_worst:<6} "
              f"{empirical_best:<6} {empirical_avg:<9.1f} {empirical_worst:<6}")

    print("\n" + "=" * 60)
    print("GROWTH ANALYSIS")
    print("=" * 60)

    print("\n1. BEST CASE ANALYSIS:")
    print("   - Theoretical: O(1) - constant time")
    print("   - Empirical: Always 1 comparison regardless of input size")
    print("   - Growth rate: No growth (constant)")

    print("\n2. AVERAGE CASE ANALYSIS:")
    print("   - Theoretical: O(n/2) - linear growth")
    print("   - Empirical results show linear growth pattern")

    if len(average_case_comparisons) > 1:
        growth_rates = []
        for i in range(1, len(average_case_comparisons)):
            growth_rate = average_case_comparisons[i] / average_case_comparisons[i-1]
            size_ratio = sizes[i] / sizes[i-1]
            growth_rates.append(growth_rate / size_ratio)

        avg_growth_rate = np.mean(growth_rates)
        print(f"   - Average growth rate factor: {avg_growth_rate:.2f}")

    print("\n3. WORST CASE ANALYSIS:")
    print("   - Theoretical: O(n) - linear growth")
    print("   - Empirical: Exactly n comparisons for each input size")
    print("   - Growth rate: Perfectly linear (1:1 ratio)")

    correlation_worst = np.corrcoef(sizes, worst_case_comparisons)[0, 1]
    correlation_avg = np.corrcoef(sizes, average_case_comparisons)[0, 1]

    print(f"\n4. CORRELATION WITH INPUT SIZE:")
    print(f"   - Worst case correlation: {correlation_worst:.4f} (perfect linear)")
    print(f"   - Average case correlation: {correlation_avg:.4f} (strong linear)")
    print(f"   - Best case correlation: 0.0000 (no correlation)")

if __name__ == "__main__":
    analyze_linear_search_comparisons()
```
