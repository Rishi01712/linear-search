```python
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, target):
    """Linear search implementation that returns the index of target if found, -1 otherwise"""
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def generate_random_list(n):
    """Generate a list of n random integers"""
    return [random.randint(1, n*10) for _ in range(n)]

def measure_search_time(arr, target, trials=1000):
    """Measure average time for searching target in array over multiple trials"""
    total_time = 0

    for _ in range(trials):
        start_time = time.time()
        linear_search(arr, target)
        end_time = time.time()
        total_time += (end_time - start_time)

    return total_time / trials

def analyze_linear_search_performance():
    """Analyze linear search performance for different cases"""

    # Test with different array sizes
    sizes = [1000, 5000, 10000, 20000, 50000]

    best_case_times = []
    average_case_times = []
    worst_case_times = []

    print("Linear Search Performance Analysis")
    print("=" * 50)

    for n in sizes:
        # Generate random array
        arr = generate_random_list(n)

        # Best case: element at first position
        best_target = arr[0]
        best_time = measure_search_time(arr, best_target)
        best_case_times.append(best_time * 1000)  # Convert to milliseconds

        # Average case: element in middle
        middle_target = arr[n//2]
        avg_time = measure_search_time(arr, middle_target)
        average_case_times.append(avg_time * 1000)

        # Worst case: element not present
        worst_target = max(arr) + 1000  # Ensure it's not in the array
        worst_time = measure_search_time(arr, worst_target)
        worst_case_times.append(worst_time * 1000)

        print(f"Array size: {n:5d}")
        print(f"  Best case (1st position):     {best_time*1000:.4f} ms")
        print(f"  Average case (middle):        {avg_time*1000:.4f} ms")
        print(f"  Worst case (not found):       {worst_time*1000:.4f} ms")
        print()

    # Plot the results
    plt.figure(figsize=(12, 8))

    plt.plot(sizes, best_case_times, marker='o', label='Best Case (O(1))', linewidth=2)
    plt.plot(sizes, average_case_times, marker='s', label='Average Case (O(n/2))', linewidth=2)
    plt.plot(sizes, worst_case_times, marker='^', label='Worst Case (O(n))', linewidth=2)

    plt.xlabel('Array Size (n)')
    plt.ylabel('Time (milliseconds)')
    plt.title('Linear Search Performance Analysis')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Add annotations
    plt.annotate('Best case remains constant\n(O(1) complexity)',
                xy=(sizes[2], best_case_times[2]),
                xytext=(sizes[2]+5000, best_case_times[2]+0.5),
                arrowprops=dict(arrowstyle='->', color='blue', alpha=0.7))

    plt.tight_layout()
    plt.show()

    # Print analysis
    print("\nTIME COMPLEXITY ANALYSIS:")
    print("=" * 50)
    print("1. BEST CASE - O(1):")
    print("   - Element found at first position")
    print("   - Time remains constant regardless of array size")
    print("   - Only one comparison needed")

    print("\n2. AVERAGE CASE - O(n/2):")
    print("   - Element found somewhere in the middle")
    print("   - On average, need to check half the elements")
    print("   - Time grows linearly with array size")

    print("\n3. WORST CASE - O(n):")
    print("   - Element not found in array")
    print("   - Must check all elements")
    print("   - Time grows linearly with array size")

    print(f"\nOBSERVATIONS:")
    print(f"- Best case time ratio (largest/smallest): {max(best_case_times)/min(best_case_times):.2f}")
    print(f"- Average case time ratio (largest/smallest): {max(average_case_times)/min(average_case_times):.2f}")
    print(f"- Worst case time ratio (largest/smallest): {max(worst_case_times)/min(worst_case_times):.2f}")
    print(f"- As input size increases, worst case time increases proportionally")

if __name__ == "__main__":
    analyze_linear_search_performance()
```
